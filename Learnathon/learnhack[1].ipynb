```python
import pandas as pd
import numpy as np
```

```python
ins=pd.read_csv(r"C:\Users\patro\Downloads\Auto_Insurance_Fraud_Claims_File03.csv")
ins
```

```python
for i in ins.columns:
    print(i,':','\n',ins[i].unique(),'\n')
```
